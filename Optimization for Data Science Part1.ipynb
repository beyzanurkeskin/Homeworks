```python
import numpy as np
import matplotlib.pyplot as plt


def f(x1, x2):                                                         # definig the function to plot
    return (x1**4)/12 - (x1**2)/2 + (x2**3)/6 + x2**2 + (3*x2)/2 + 20


x = np.linspace(-3, 3, 100)                                            # definig the interval to plot the surface
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)


D = f(X, Y)                                                     # calculating the function values for each point in the grid


levels = np.linspace(20, 40, 21)                                      # defini the levels for the contour plot


plt.contour(X, Y, D, levels=levels)                                   # creatng the contour plot

U, V = np.gradient(D)                                                 # calculating the gradient of the function to
U = U / np.sqrt(U**2 + V**2)                                       #  plot arrows representing the direction of the gradient
V = V / np.sqrt(U**2 + V**2)


plt.quiver(X, Y, U, V)                                               # plotng the arrows representing the gradient


plt.show()
```
